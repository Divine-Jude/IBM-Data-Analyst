```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set()

QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df.hist(column='ConvertedComp')

plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.title('Histogram of Converted Compensation')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df.boxplot(column = 'Age')

plt.title('Box plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""

df = pd.read_sql(QUERY, conn)

sns.scatterplot(data=df, x='Age', y='WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT * FROM master
"""
df1 = pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=df1, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=1, sizes=(100, 900)) 

plt.title('WorkWeekHrs and CodeRevHrs By Age', size=14) 
plt.xlabel('WorkWeekHrs', size=10) 
plt.ylabel('CodeRevHrs', size=10)

plt.show()


```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT * FROM DatabaseDesireNextYear
"""
df2 = pd.read_sql_query(QUERY,conn)
db_counts = df2['DatabaseDesireNextYear'].value_counts()
top_5_db = db_counts.head(5)
print(top_5_db)


plt.pie(top_5_db, labels=top_5_db.index, autopct='%.1f%%')
plt.title('Top 5 Databases That Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)
df1 = df.groupby('Age').median()
# print(df1)

df1.plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('WorkWeekHrs and CodeRevHrs')
plt.ylabel('Median')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)
df1 = df.groupby('Age').median()
# print(df1)

df1.plot(kind='line', figsize=(10, 6), stacked=True)
plt.title('Median Converted Compensation for Age Group 45-60')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation')
plt.show()
```

Based on the Line Chart of Age and median ConvertedComp for the age group 25-30, what kind of correlation do you see between Age and median ConvertedComp? 

```python
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 20 AND 35
"""
df = pd.read_sql_query(QUERY, conn)
df1 = df.groupby('Age').median()
# print(df1)

df1.plot(kind='line', figsize=(10, 6), stacked=True)
plt.title('Median Converted Compensation for Age Group 45-60')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch FROM master
"""

df = pd.read_sql_query(QUERY, conn)
main_branch_counts = df['MainBranch'].value_counts()
# print(main_branch_counts)

plt.barh(y=main_branch_counts.index, width=main_branch_counts.values)
for i, v in enumerate(main_branch_counts.values):
    plt.text(v, i, v)
plt.title('Main Branch Counts')
plt.xlabel('Counts')
plt.ylabel('Main Branch')
plt.show()

```
